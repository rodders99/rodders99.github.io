```python
def top3s(country, frame):
    df = frame
    df_result = df[df['country']==country]

    df_risk = df_result[df_result['type'] == 'risk'].copy()
    df_risk['image'] = df_risk['column'].map(img_dict)
    df_risk = df_risk.replace(' ', '\n', regex=True)
    df_risk['percentage'] = df_risk['percentage'].round(0).astype(int).astype(str) +' %'
    

    df_cat = df_result[df_result['type'] == 'category'].copy()
    df_cat['image'] = df_cat['column'].map(img_dict)
    
    df_cat = df_cat.replace(' ', '\n', regex=True)
    df_cat['percentage'] = df_cat['percentage'].round(0).astype(int).astype(str) + " %"
    

    country = df_cat['country'].iloc[0]
    
    


    import matplotlib.pyplot as plt
    import seaborn as sns
    from matplotlib.offsetbox import OffsetImage, AnnotationBbox
    import matplotlib.image as mpimg
    
    sns.set_theme(style="darkgrid")
    plt.style.use('fivethirtyeight')
    fig, ax = plt.subplots(figsize=(12, 4))
    
    color1 = '#4a94c9'
    color2 = '#ec8700'
    color3 = '#dbeaf4'
    color4 = '#fbe7cc'
    bgcolor = '#aecfe7'
    
    
    fig.patch.set_facecolor(bgcolor)
    ax.set_facecolor('none')
    
    #plt.suptitle("2023", fontsize=16, ha='center', color="black")  # Adjust y position as needed
    
    ax.text(0.04, 0.7, df_cat['country'].iloc[0], va = 'top',ha='center', fontsize=14, color='grey')
    
    # Adding Section Titles
    
    ax.text(0.33, 0.9, 'THE THREE MOST COMMON ', ha='center', fontsize=18, color='white')
    ax.text(0.33, 0.8, 'PRODUCT CATEGORIES NOTIFIED', ha='center',fontsize=18, color='white')
    # Draw section HEADER rectangle
    p_box = plt.Rectangle((0.11, 0.76), 0.44, 0.23, color=color1, fill=color1, zorder = -1)
    # Draw section CONTENT rectangle
    p_lightbox = plt.Rectangle((0.11, 0.02), 0.44, 0.768, color=color3, fill=color1, zorder = -1)
    
    
    
    df = df_cat
    
    image_path = 'img/icons/' + df['image'].iloc[0]  # Update this to your image's path
    cat_image0 = mpimg.imread(image_path)
    cat_offset_image0 = OffsetImage(cat_image0, zoom=0.09)
    cat_image_box0 = AnnotationBbox(cat_offset_image0, (0.172, 0.45), frameon=False)
    
    image_path = 'img/icons/' + df['image'].iloc[1]  # Update this to your image's path
    cat_image1 = mpimg.imread(image_path)
    cat_offset_image1 = OffsetImage(cat_image1, zoom=0.09)
    cat_image_box1 = AnnotationBbox(cat_offset_image1, (0.32, 0.45), frameon=False)
    
    image_path = 'img/icons/' + df['image'].iloc[2]  # Update this to your image's path
    cat_image2 = mpimg.imread(image_path)
    cat_offset_image2 = OffsetImage(cat_image2, zoom=0.09)
    cat_image_box2 = AnnotationBbox(cat_offset_image2, (0.475, 0.45), frameon=False)
    
    
    ax.text(0.175, 0.7, df['percentage'].iloc[0], va = 'top', ha='center', fontsize=14, color=color1, fontweight='bold')
    #ax.text(0.175, 0.5, 'ICON1', ha='center', fontsize=14, color=color1)
    ax.text(0.175, 0.25, df['column'].iloc[0], va = 'top', ha='center', fontsize=14, color=color1, fontweight='bold')
    
    
    ax.text(0.32, 0.7, df['percentage'].iloc[1], va = 'top', ha='center', fontsize=14, color=color1, fontweight='bold')
    #ax.text(0.32, 0.5, 'ICON2', ha='center', fontsize=14, color=color1)
    ax.text(0.32, 0.25, df['column'].iloc[1], va = 'top', ha='center', fontsize=14, color=color1, fontweight='bold')
    
    ax.text(0.48, 0.7, df['percentage'].iloc[2], va = 'top', ha='center', fontsize=14, color=color1, fontweight='bold')
    #ax.text(0.48, 0.5, 'ICON3', ha='center', fontsize=14, color=color1)
    ax.text(0.48, 0.25, df['column'].iloc[2], va = 'top', ha='center', fontsize=14, color=color1, fontweight='bold')
    
    #####################################
    
    # Adding Section Titles
    
    ax.text(0.79, 0.9, 'THE THREE MOST COMMON ', ha='center', fontsize=18, color='white')
    ax.text(0.79, 0.8, 'RISKS NOTIFIED', ha='center',fontsize=18, color='white')
    
    # Draw section HEADER rectangle
    r_box = plt.Rectangle((0.58, 0.76), 0.44, 0.23, color=color2, fill=color2, zorder = -1)
    # Draw section CONTENT rectangle
    r_lightbox = plt.Rectangle((0.58, 0.02), 0.44, 0.768, color=color4, fill=color1, zorder = -1)
    
    
    df = df_risk
    
    image_path = 'img/icons/' + df['image'].iloc[0]  # Update this to your image's path
    
    risk_image0 = mpimg.imread(image_path)
    risk_offset_image0 = OffsetImage(risk_image0, zoom=0.09)
    risk_image_box0 = AnnotationBbox(risk_offset_image0, (0.64, 0.45), frameon=False)
    
    image_path = 'img/icons/' + df['image'].iloc[1]  # Update this to your image's path
    risk_image1 = mpimg.imread(image_path)
    risk_offset_image1 = OffsetImage(risk_image1, zoom=0.09)
    risk_image_box1 = AnnotationBbox(risk_offset_image1, (0.785, 0.45), frameon=False)
    
    image_path = 'img/icons/' + df['image'].iloc[2]  # Update this to your image's path
    risk_image2 = mpimg.imread(image_path)
    risk_offset_image2 = OffsetImage(risk_image2, zoom=0.09)
    risk_image_box2 = AnnotationBbox(risk_offset_image2, (0.93, 0.45), frameon=False)
    
    
    ax.text(0.645, 0.7, df['percentage'].iloc[0], va = 'top', ha='center', fontsize=14, color=color2, fontweight='bold')
    #ax.text(0.645, 0.5, 'ICON1', ha='center', fontsize=14, color=color2)
    ax.text(0.645, 0.25, df['column'].iloc[0], va = 'top', ha='center', fontsize=14, color=color2, fontweight='bold')
    
    ax.text(0.79, 0.7, df['percentage'].iloc[1], va = 'top', ha='center', fontsize=14, color=color2, fontweight='bold')
    #ax.text(0.79, 0.5, 'ICON2', ha='center', fontsize=14, color=color2)
    ax.text(0.79, 0.25, df['column'].iloc[1], va = 'top', ha='center', fontsize=14, color=color2, fontweight='bold')
    
    ax.text(0.935, 0.7, df['percentage'].iloc[2], va = 'top', ha='center', fontsize=14, color=color2, fontweight='bold')
    #ax.text(0.935, 0.5, 'ICON2', ha='center', fontsize=14, color=color2)
    ax.text(0.935, 0.25, df['column'].iloc[2], va = 'top', ha='center', fontsize=14, color=color2, fontweight='bold')
    
    # Drawing a circle
    
    ax.add_artist(p_box)
    ax.add_artist(p_lightbox)
    ax.add_artist(r_box)
    ax.add_artist(r_lightbox)
    
    ax.add_artist(cat_image_box0)
    ax.add_artist(cat_image_box1)
    ax.add_artist(cat_image_box2)
    ax.add_artist(risk_image_box0)
    ax.add_artist(risk_image_box1)
    ax.add_artist(risk_image_box2)
    
    ax.grid(False)
    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    
    ax.set_xticks([])
    ax.set_yticks([])
    
    
    # Setting limits and showing plot
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    plt.tight_layout()
    plt.show()
```
